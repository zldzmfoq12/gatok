```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
import time
#import pymysql

#conn = pymysql.connect(host='localhost', user='hyhci', password='ashia19969', db='politicians', charset='utf8')

#curs = conn.cursor()

photo = []
name = []
birth = []
hometown = []
party = []
span = []
span.append(1)

driver = webdriver.Chrome('/Applications/chromedriver')
driver.set_window_size(1440,1080)
driver.implicitly_wait(3)
driver.get('https://en.wikipedia.org/wiki/List_of_current_members_of_the_United_States_House_of_Representatives')
driver.implicitly_wait(3)
    
def politicians(i):
    try:
        if span[-1] > 1:
            span.append(span[-1]-1)
            
            party_num = party[-1]
            party.append(party_num)
            
            name_in = driver.find_element_by_css_selector('#votingmembers > tbody > tr:nth-child('+str(i+1) +') > td:nth-child(2) > span > span > span > a')
            name_text = name_in.text
            name.append(name_text)
            print('---'+name_text+'--- 정보 처리')

            photo_in = driver.find_element_by_css_selector('#votingmembers > tbody > tr:nth-child('+str(i+1) +') > td:nth-child(2) > a > img')
            photo_url = photo_in.get_attribute('src')
            photo.append(photo_url)

            birth_in = driver.find_element_by_css_selector('#votingmembers > tbody > tr:nth-child('+str(i+1) +') > td:nth-child(7)')
            birth_text = birth_in.text
            birth.append(birth_text)
        else:
            span_in = driver.find_element_by_css_selector('#votingmembers > tbody > tr:nth-child('+str(i+1) +') > td:nth-child(4)')
            span_num = span_in.get_attribute('rowspan')
            if span_num == None:
                span.append(1)
            else:
                span.append(int(span_num))
            
            party_in = driver.find_element_by_css_selector('#votingmembers > tbody > tr:nth-child('+str(i+1) +') > td:nth-child(4)')
            party_text = party_in.text
            if party_text == 'Republican':
                party_num = 1
            elif party_text == 'Democratic':
                party_num = 0
            else:
                party_num = 2
            party.append(party_num)

            name_in = driver.find_element_by_css_selector('#votingmembers > tbody > tr:nth-child('+str(i+1) +') > td:nth-child(2) > span > span > span > a')
            name_text = name_in.text
            name.append(name_text)
            print('---'+name_text+'--- 정보 처리')

            photo_in = driver.find_element_by_css_selector('#votingmembers > tbody > tr:nth-child('+str(i+1) +') > td:nth-child(2) > a > img')
            photo_url = photo_in.get_attribute('src')
            photo.append(photo_url)

            birth_in = driver.find_element_by_css_selector('#votingmembers > tbody > tr:nth-child('+str(i+1) +') > td:nth-child(9)')
            birth_text = birth_in.text
            birth.append(birth_text)
            
    except NoSuchElementException:
        pass
    
for i in range(0, 435):
    if i == 330:
        pass
    else:
        politicians(i)
        time.sleep(1)
print("---정보 처리 끝---")
```

```python
import pandas as pd

df = pd.DataFrame({'name':name, 'photo':photo, 'age':birth, 'party':party})
```

```python
df.to_csv('./representative.csv')
```
